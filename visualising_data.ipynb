```python
#data = pd.read_csv("03. Avg amplitude values.csv")
#data = pd.read_csv("04. Avg frequency values.csv")
#data = pd.read_csv("05. Avg syllable duration values.csv")
data = pd.read_csv("06. Avg internote interval values.csv")
```

```python
sns.set_style("ticks")
plt.figure(dpi=300)
colors = ["#3274a1", "#e1812c", "#3a923a", "#c03d3e"]
species = ['BCC', 'BF', 'JF', 'ZF']
x = [1, 2, 3, 4, 5]

for j in range(len(species)):
    df1 = data.loc[data['Species'] == species[j]]
    df2 = df1.iloc[:, 2:]
    df_mean = np.array(np.mean(df2))
    df_sd = np.array(np.std(df2))
    for i in range(len(df1)-1):
        p = df1.iloc[i:i+1, :]
        q = np.array(p.iloc[:, 2:])
        y = q[0]
        plt.plot(x, y, linewidth=0.75, color=colors[j])
    plt.plot(x, df_mean, linewidth=1.5, color=colors[j], linestyle='dashed', label=species[j])

plt.legend(bbox_to_anchor=(1.2, 0.3))
#plt.ylim(top=50)

plt.xlabel('Syllable position', fontsize = 12)
plt.xticks(np.arange(1, 6, step=1))
plt.ylabel('Log Amplitude (dB)', fontsize = 12)
plt.savefig('08 Amplitude IN all birds.png', dpi=600)
#plt.show()
```

```python
sns.set_style("ticks")
plt.figure(dpi=300)
colors = ["#3274a1", "#e1812c", "#3a923a", "#c03d3e"]
species = ['BCC', 'BF', 'JF', 'ZF']
x = [1, 2, 3, 4, 5]

for j in range(len(species)):
    df1 = data.loc[data['Species'] == species[j]]
    df2 = df1.iloc[:, 2:]
    df_mean = np.array(np.mean(df2))
    df_sd = np.array(np.std(df2))
    df_sem = np.array(np.std(df2)/np.sqrt(len(df1)))
    # for i in range(len(df1)-1):
    #     p = df1.iloc[i:i+1, :]
    #     q = np.array(p.iloc[:, 2:])
    #     y = q[0]
    #     plt.plot(x, y, linewidth=0.75, color=colors[j])
    plt.plot(x, df_mean, linewidth=1.5, color=colors[j], linestyle='dashed', label=species[j])
    plt.fill_between(x, df_mean-df_sem, df_mean+df_sem, color=colors[j], alpha=0.5)

#plt.legend(bbox_to_anchor=(1.2, 0.3))
#plt.ylim(top=50)

plt.xlabel('Syllable position', fontsize = 12)
plt.xticks(np.arange(1, 6, step=1))
plt.ylabel('Internote interval (ms)', fontsize = 12)
plt.savefig("11 Internote interval IN.jpg", dpi=600)
#plt.show()
```

```python
data = pd.read_csv("07. Start-Self Transition Prob.csv")
```

```python
sns.set_style("ticks")
plt.figure(dpi=600)
#g = sns.scatterplot(data=data, x="Start transition prob", y="Self-transition prob", hue="Species")
#sns.swarmplot(data=song2_q, x="Species", y="Avg. repeats of first syll", color="k", size=5, ax=g.ax)
g = sns.jointplot(data=data, x="Start transition prob", y="Self-transition prob", kind="scatter", marker="D", 
    hue="Species", xlim=(-0.05, 1.05), ylim=(-0.05, 1.05))

plt.xlabel('Beginning bout probability', fontsize = 14)
plt.ylabel('Self-transition probability', fontsize = 14)

#plt.setp(g.get_legend().get_title(), fontsize='6')
#g.legend(bbox_to_anchor=(1, 1))
#plt.show()
plt.savefig('12 Start and Self transition prob.png', dpi = 600)
```

```python
from IPython.display import Image

fig = px.scatter(data, x="Start transition prob", y="Self-transition prob", template="plotly_white",
    color="Species", marginal_x="histogram", marginal_y="histogram", color_discrete_sequence = ["#3274a1", "#e1812c", "#3a923a", "#c03d3e"])



fig.update_traces(
    marker=dict(size=7, symbol="diamond", line=dict(width=1, color="DarkSlateGrey")),
    selector=dict(mode="markers"),
)

#fig.write_image("Start-Self trans.png")
img_bytes = fig.to_image(format="png", width=1200, height=1200, scale=5)
Image(img_bytes)

fig.write_image("Start-Self trans.png")

fig.show()
```
