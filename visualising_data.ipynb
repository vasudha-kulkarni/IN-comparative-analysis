```python
df = pandas.read_csv('06. Avg internote interval values.csv')
print(df.columns)

data = df.iloc[:, 2:]
data = np.array(data)
```

```python
x = [1, 2, 3, 4, 5]
#colors = {'BCC' : 'tab:blue', 'BF' : 'tab:brown', 'JF' : 'tab:grey', 'ZF' : 'tab:green'}
#c=df['Species'].map(colors)

for i in range(len(df)):
    y = data[i]
    plt.plot(x, y, linewidth = 0.5)

plt.xlabel('Syllable position', fontsize = 12)
plt.ylabel('Internote interval (ms)', fontsize = 12)
plt.title('Duration of interval between initial syllables', fontsize = 14)

plt.show()
plt.savefig('04 internote interval init syll.png', dpi = 600)
```
