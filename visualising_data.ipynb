```python
import math
import pandas
import os
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pandas.read_csv('syllable_data.csv')
print(df.columns)
```

```python
sns.set()

fig, ax = plt.subplots()
sns.swarmplot('Species', '#Syl', data=df)
#ax.legend(loc='upper left', fontsize = 20, bbox_to_anchor = (0, 1.1))
ax.set_xlabel('Species', fontsize = 12)
ax.set_ylabel('Number of syllables', fontsize = 12)
ax.set_title('No. of unique syllables across species', fontsize = 16)

#plt.show()
plt.savefig('01 Num syllables.png', dpi = 600)
```
