```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import umap
from sklearn.manifold import TSNE
import plotly.express as px
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
import os
```

```python
data = pd.read_csv("/content/BF_brn24pnk13.txt", sep='\t', header=None, 
       names = ['FileName', 'Syll #', 'SyllLabel',
       ' Syll Onset (ms)', ' Syll Offset (ms)', ' Syll Duration (sec)', ' Mean Frequency (Hz)',
       ' Entropy', 'Log Amplitude (dB)', 'Pitch Goodness', 'FrequencyModulation',
       'AmplitudeModulation', 'EntropyVariance', 'Fundamental Frequency (Hz)', 'RMS Amplitude'])
data = data.iloc[1:, :]
```

```python
data1 = data.loc[:, ' Syll Duration (sec)':'EntropyVariance']
data1 = data1.astype(float)
```

```python
tsne = TSNE(n_components=2, random_state=0)
projections = tsne.fit_transform(data1)

fig = px.scatter(
    projections, x=0, y=1,
    color = data.SyllLabel, labels={'color': 'SyllLabel'}
)
fig.show()
```

```python
umap_2d = umap.UMAP(n_components=2, init='random', random_state=0)
#umap_3d = UMAP(n_components=3, init='random', random_state=0)

proj_2d = umap_2d.fit_transform(data1)
#proj_3d = umap_3d.fit_transform(features)

fig_2d = px.scatter(
    proj_2d, x=0, y=1,
    color=data.SyllLabel, labels={'color': 'SyllLabel'}
)

fig_2d.show()
```
